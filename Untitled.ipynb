```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten

# 建立簡單的線性執行的模型
model_neural = Sequential()
# Add Input layer
model_neural.add(Dense(units=512, input_dim=784, kernel_initializer='normal', activation='relu'))
# 中間的hidden layer
model_neural.add(Dense(units=512, input_dim=512, kernel_initializer='normal', activation='relu')) 
model_neural.add(Dense(units=256, input_dim=512, kernel_initializer='normal', activation='relu'))
# Add output layer
model_neural.add(Dense(units=10, kernel_initializer='normal', activation='softmax'))

##loss function:cross-entropy
##Optimizer是要調整參數，來使Loss越小越好:選
model_neural.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 

# 進行訓練, 訓練過程會存在 train_history 變數中
train_history = model_neural.fit(x=x_train, y=y_train_dummy, validation_split=0.2, epochs=10, batch_size=800, verbose=2)  

# 顯示訓練成果(分數)
scores = model_neural.evaluate(x_test, y_test_dummy)  
print()  
print("\t[Info] Accuracy of testing data = {:2.1f}%".format(scores[1]*100.0))  

# 預測(prediction)
X = x_test[:,:]
predictions = np.argmax(model_neural.predict(X), axis=-1)
# get prediction result
print(predictions)
```

```python
from torch.nn import Module
from torch import nn
import torch.nn.functional as F

class Lenet_Model(nn.Module):
    def __init__(self):
        super(Lenet_Model, self).__init__()
        
        ##torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, 
        ##padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros', 
        ##device=None, dtype=None)
        
        
        
        ##6 kernels of size 5×5 and the stride of 1. 
        ##Given the input size (32×32×1), the output of this layer is of size 28×28×6.
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1)
        ##activate fun: relu
        self.relu1 = nn.ReLU()
        ##pooling layer with 6 kernels of size 2×2, stride of 2. 
        self.pool1 = nn.MaxPool2d(2)
        ##
        self.conv2 = nn.Conv2d(6, 16, 5)
        ##activate fun: relu
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(2)
        self.fc1 = nn.Linear(256, 120)
        ##activate fun: relu
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(120, 84)
        ##activate fun: relu
        self.relu4 = nn.ReLU()
        self.fc3 = nn.Linear(84, 10)
        self.relu5 = nn.ReLU()

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.pool1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.pool2(x)
        x = x.view(x.shape[0], -1)
        x = self.fc1(x)
        x = self.relu3(x)
        x = self.fc2(x)
        x = self.relu4(x)
        x = self.fc3(x)
        x = self.relu5(x)
        return x
    
model_lenet = Lenet_Model()
print(model_lenet)
```

```python
from torchvision import datasets
import torchvision.transforms as transforms


# convert data to torch.FloatTensor
transform = transforms.ToTensor()
train_data = datasets.MNIST(root='data', train=True,download=True, transform=transform)
test_data = datasets.MNIST(root='data', train=False,download=True, transform=transform)

# prepare data loaders
train_loader = torch.utils.data.DataLoader(train_data, batch_size=20,num_workers=0)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=20,num_workers=0)
```

```python
## Specify loss and optimization functions

##1. specify loss function
criterion = nn.CrossEntropyLoss()

##2. specify optimizer
optimizer = torch.optim.SGD(model_lenet.parameters(), lr=0.01)

# number of epochs to train the model
n_epochs = 10  # suggest training between 20-50 epochs

model_lenet.train() # prep model for training

for epoch in range(n_epochs):
    # monitor training loss
    train_loss = 0.0
    
    ###################
    # train the model #
    ###################
    for data, target in train_loader:
        # clear the gradients of all optimized variables
        optimizer.zero_grad()
        # forward pass: compute predicted outputs by passing inputs to the model
        output = model_lenet(data)
        # calculate the loss
        loss = criterion(output, target)
        # backward pass: compute gradient of the loss with respect to model parameters
        loss.backward()
        # perform a single optimization step (parameter update)
        optimizer.step()
        # update running training loss
        train_loss += loss.item()*data.size(0)
        
    # print training statistics 
    # calculate average loss over an epoch
    train_loss = train_loss/len(train_loader.dataset)

    print('Epoch: {} \tTraining Loss: {:.6f}'.format(
        epoch+1, 
        train_loss
        ))
```
